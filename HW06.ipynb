## Question 1

The theoretical Simple Linear Regression model can estimate the dependent variable with the predictor. The relationship is $Y_i$ = $\beta_0$ + $\beta_1$ $x_i$ + $\epsilon_i$

- $Y_i$ is the outcome we want to estimate with the simple linear regression model.
- $x_i$ is the predictor to estimate $Y_i$
- $\beta_1$ is the slope coefficient, which represent the change of $Y_i$ for one unit change of $x_i$ in the simple linear regression model.
- $\beta_0$ is the intercept of the linear regression model, represent the $Y_i$ by estimating when $x_i$ = 0
- $\epsilon_i$ is the errors in a theoretical simple linear regression model, which makes simple linear regression a statistical model by introducing a radom variable with a distribution.

- The $Y_i$ = $\beta_0$ + $\beta_1$ $x_i$ + $\epsilon_i$ represent a theoretical model which give the relationship between the dependente value $Y_i$ and $x_i$.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Set a random seed for reproducibility
np.random.seed(42)

# Define the theoretical parameters
beta_0 = 2  # Intercept
beta_1 = 3  # Slope
sigma = 1   # Standard deviation of errors (ϵ_i)

# Generate a sample dataset
n = 50  # Number of data points
x = np.linspace(0, 10, n)  # Predictor variable x_i

# Generate errors from a normal distribution
epsilon = norm.rvs(loc=0, scale=sigma, size=n)

# Calculate Y_i based on the theoretical model Y_i = β_0 + β_1 * x_i + ε_i
y = beta_0 + beta_1 * x + epsilon

# Calculate the theoretical values of y without the error term
y_theoretical = beta_0 + beta_1 * x

# Plot observed data points, the true regression line, and the individual errors (epsilon)
plt.figure(figsize=(10, 6))

# Plot observed data points
plt.scatter(x, y, color='blue', label='Observed data points')

# Plot the true regression line
plt.plot(x, y_theoretical, color='red', linestyle='--', label='True regression line (Y = 2 + 3x)')

# Plot vertical lines to show individual errors (epsilon)
for i in range(n):
    plt.vlines(x[i], y[i], y_theoretical[i], color='gray', linestyle=':', alpha=0.7)

# Labels and legend
plt.xlabel('x')
plt.ylabel('y')
plt.title('Theoretical Simple Linear Regression Model with Errors (ε)')
plt.legend()
plt.show()
```

## Question 2

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
from scipy.stats import norm

# Set a random seed for reproducibility
np.random.seed(42)

# Define the theoretical parameters
beta_0 = 2  # Intercept
beta_1 = 3  # Slope
sigma = 1   # Standard deviation of errors (ϵ_i)

# Generate a sample dataset
n = 50  # Number of data points
x = np.linspace(0, 10, n)  # Predictor variable x_i

# Generate errors from a normal distribution
epsilon = norm.rvs(loc=0, scale=sigma, size=n)

# Calculate Y_i based on the theoretical model Y_i = β_0 + β_1 * x_i + ε_i
y = beta_0 + beta_1 * x + epsilon

# Create a DataFrame to hold the simulated data
data = pd.DataFrame({'x': x, 'y': y})

# Fit a Simple Linear Regression model using statsmodels
model = smf.ols('y ~ x', data=data).fit()

# Extract fitted values (predicted y) for visualization
data['fitted_y'] = model.fittedvalues

# Plot observed data points, the true regression line, and the fitted regression line
plt.figure(figsize=(10, 6))

# Plot observed data points
plt.scatter(data['x'], data['y'], color='blue', label='Observed data points')

# Plot the true regression line
plt.plot(data['x'], beta_0 + beta_1 * data['x'], color='red', linestyle='--', label='True regression line (Y = 2 + 3x)')

# Plot the fitted regression line
plt.plot(data['x'], data['fitted_y'], color='green', label='Fitted regression line')

# Labels and legend
plt.xlabel('x')
plt.ylabel('y')
plt.title('Theoretical vs Fitted Simple Linear Regression Model')
plt.legend()
plt.show()
```

## Question 9

- null hypothesis: there is a relationship between duration and wait time 
- alternative hypothesis: there is not a relationship between duration and wait time 

### short wait times which are less than short_wait_limit values of 62

```python
import seaborn as sns
import statsmodels.formula.api as smf

# The "Classic" Old Faithful Geyser dataset
old_faithful = sns.load_dataset('geyser')

linear_for_specification = 'duration ~ waiting'
model = smf.ols(linear_for_specification, data=old_faithful)
fitted_model = model.fit()
fitted_model.summary()
import plotly.express as px
import statsmodels.formula.api as smf


short_wait_limit = 62 # 64 # 66 #
short_wait = old_faithful.waiting < short_wait_limit

print(smf.ols('duration ~ waiting', data=old_faithful[short_wait]).fit().summary().tables[1])

# Create a scatter plot with a linear regression trendline
fig = px.scatter(old_faithful[short_wait], x='waiting', y='duration', 
                 title="Old Faithful Geyser Eruptions for short wait times (<"+str(short_wait_limit)+")", 
                 trendline='ols')

fig.show() # USE `fig.show(renderer="png")` FOR ALL GitHub and MarkUs SUBMISSIONS
```

p-value = 0.238, therefore we have No evidence against the null hypothesis, and reject the alternative hypothesis. hence, we cannot conclude that there is not a relationship between duration and wait time.
